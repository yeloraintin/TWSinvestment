```python
import pandas as pd
import requests
import numpy as np
from io import StringIO
import json
import matplotlib.pyplot as plt
```

```python
def craw_one_month(date):
    url='https://www.twse.com.tw/exchangeReport/BWIBBU_d?response=json&date='+format(date)+'&selectType=ALL'
    info=requests.get(url)
    data = json.loads(info.text)
    return pd.DataFrame(data['data'],columns=data['fields'])
```

```python
pd.set_option('display.max_rows', None)
x=craw_one_month(20210205)
x
```
