```python
import requests
import pandas as pd
```

```python
# 爬取目標網站
year = 109
season = 3
stock_number = 2330

#資產負債表
BalanceSheetURL = "https://mops.twse.com.tw/mops/web/t164sb03"

#損益表
ProfitAndLoseURL = "https://mops.twse.com.tw/mops/web/t164sb04"

#現金流量表
CashFlowStatementURL = "https://mops.twse.com.tw/mops/web/t164sb05"
```

```python
def crawl_financial_Report(url):
    
    form_data = {
        'encodeURIComponent':1,
        'step':1,
        'firstin':1,
        'off':1,
        'co_id':stock_number,
        'year': year,
        'season': season,
    }

    r = requests.post(url,form_data)
    html_df = pd.read_html(r.text)[10].fillna("")
    return html_df
```

```python
#資產負債表
pd.set_option('display.max_rows', None)
crawl_financial_Report(BalanceSheetURL)
```

```python
#損益表
crawl_financial_Report(ProfitAndLoseURL)
```

```python
#現金流量表
crawl_financial_Report(CashFlowStatementURL)
```
